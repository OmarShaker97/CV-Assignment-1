```python
import matplotlib.pyplot as plt
from PIL import Image

img = Image.open("Cameraman_noise.bmp").convert("L")
 
averaging_img = img.copy()

averaging_value = 0
 
# CALCULATE FOR ALL PIXELS THAT ARE ON THE LEFT    
for i in range (0,img.size[0]):
    for j in range (0,img.size[1]):
        try:
            top_pixel = img.getpixel((i, j-1))
        except:
            top_pixel = 0
            
        try:
            bottom_pixel = img.getpixel((i,j+1))
        except:
            bottom_pixel = 0
            
        try:
            right_pixel = img.getpixel((i+1, j))
        except:
            right_pixel = 0
            
        try:
            left_pixel = img.getpixel((i-1,j))
        except:
            left_pixel = 0
            
        try:
            top_left_pixel = img.getpixel((i-1, j-1))
        except:
            top_left_pixel = 0
       
        try:
            top_right_pixel = img.getpixel((i+1, j-1))    
        except:
            top_right_pixel = 0
            
        
        try:
            bottom_right_pixel = img.getpixel((i+1, j+1))
        except:
            bottom_right_pixel = 0
      
        try:
            bottom_left_pixel = img.getpixel((i-1,j+1))
        except:
            bottom_left_pixel = 0
            
        averaging_value = (img.getpixel((i,j)) + top_pixel + bottom_pixel + left_pixel + right_pixel + top_left_pixel + top_right_pixel + bottom_right_pixel + bottom_left_pixel)//9
        
        averaging_img.putpixel((i,j),averaging_value)
        
        
       
plt.imshow(averaging_img)
averaging_img.save('Camera_Filt_3.jpg')
```
