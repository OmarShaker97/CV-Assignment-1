```python
def change_contrast(A, B, C, D, out_image_name): 
    img = Image.open("Ocean.bmp").convert('L')
    gray_img = img.copy()
    for i in range (0,img.size[0]):
        for j in range (0,img.size[1]):
            currPixel= img.getpixel((i, j))
            m = 0
            b = 0
            if(currPixel <= A):
                m = (B-0)/(A-0)
                b = B - int(m * A)
            elif(currPixel <= C):
                m = (D-B)/(C-A)
                b = D - int(m * C)
                #print(b)
            else:
                m = (255-D)//(255-C)
                b = 255 - int(m * 255)
            resultedPixel = int(m * (currPixel)) + b
            gray_img.putpixel((i,j),(resultedPixel))
    plt.imshow(gray_img, cmap="gray")
    gray_img.save(out_image_name)
```

```python
import numpy as np
file1 = open("contrast.txt","w") 
arr = np.zeros((256, 256))
sum = 0
img = Image.open("Ocean_a.bmp")
gray_img = img.convert('L')
for i in range (0,gray_img.size[0]):
    for j in range (0,gray_img.size[1] - 1):
        currPixel = gray_img.getpixel((i, j))
        currPixel1 = gray_img.getpixel((i, j + 1))
        arr[currPixel,currPixel1] = arr[currPixel,currPixel1] + 1;
for i in range (0,arr.shape[0]):
    for j in range (0,arr.shape[1]):
        sum += arr[i,j]*np.power((i-j), 2)
file1.write("Ocean_a.bmp: "+str(sum))
file1.write("\n")
print(sum)
arr = np.zeros((256, 256))
sum = 0
img = Image.open("Ocean_b.bmp")
gray_img = img.convert('L')
for i in range (0,gray_img.size[0]):
    for j in range (0,gray_img.size[1] - 1):
        currPixel = gray_img.getpixel((i, j))
        currPixel1 = gray_img.getpixel((i, j + 1))
        arr[currPixel,currPixel1] = arr[currPixel,currPixel1] + 1;
for i in range (0,arr.shape[0]):
    for j in range (0,arr.shape[1]):
        sum += arr[i,j]*np.power((i-j), 2)
file1.write("Ocean_b.bmp: "+str(sum))
print(sum)
file1.close()
```
